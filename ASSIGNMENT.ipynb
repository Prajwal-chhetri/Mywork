```python
import pandas as pd
```

```python
#create a two tables 
t2 = pd.DataFrame({"SHOPID":  [100, 101, 102, 103],  "SHOPNAME":["Zudio","Tata","Reliance",""],  "SHOP_LOCATION": ["Baggy_Jeans", "Earbuds", "Nexon", "HP_Laptop"], 
"SALES": [10000, 30000, 15000, 50000]})
t2
```

```python
#create table 2
t3 = pd.DataFrame({"PRODUCTID": [999, 994, 990, 991],  "SHOPID": [100, 101, 102, 104],  "PRODUCTNAME": ["Baggy_Jeans", "Earbuds", "Nexon", "HP_Laptop"], 
"PRODUCTPRICE": [2000, 3000, 1500, 5000]}, index=[500, 600, 700, 800])
t3
```

```python
# join the tables on the  SHOPID 

join_table = pd.merge(t2,t3, on="SHOPID", how="inner")
print(join_table)
```

```python
join_table = pd.merge(t2,t3, on="SHOPID", how="left")
print(join_table)
```

```python
join_table = pd.merge(t2,t3, on="SHOPID", how="right")
print(join_table)
```

```python
join_table = pd.merge(t2,t3, on="SHOPID", how="outer")
print(join_table)
```
